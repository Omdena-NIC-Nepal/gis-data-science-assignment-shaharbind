# Importing Libraries

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import rasterio
from rasterio.plot import show
from rasterstats import zonal_stats
```

# Loading Nepal Map

```python
# Load Nepal map shapefile
nepal_map = gpd.read_file('data/Shape_Data/local_unit.shp')

# Convert CRS to EPSG:4326 for consistency
nepal_map = nepal_map.to_crs(epsg=4326)

# Display first few rows
print(nepal_map.head())

# Check available province names
print(nepal_map['Province'].unique())
```

```python
# Rename province names for consistency
nepal_map['Province'] = nepal_map['Province'].replace({
    'Province 1': 'Koshi Province',
    'Province 2': 'Madhesh Province',
    'Bagmati': 'Bagmati Province',
    'Gandaki': 'Gandaki Province',
    'Province 5': 'Lumbini Province',
    'Karnali': 'Karnali Province',
    'Sudur Pashchim': 'Sudurpashchim Province'
})

# Verify changes
print(nepal_map['Province'].unique())
```

#  Plotting Nepal Map by Province

```python
fig, ax = plt.subplots(figsize=(12, 8))
nepal_map.plot(column='Province', cmap='viridis', legend=True, ax=ax, edgecolor='black', linewidth=0.2)
ax.set_title('Nepal Administrative Map')
plt.show()
```

# Loading Nepal Climate Data

```python
# Load climate data (administrative regions)
admin_regions = gpd.read_file('data/nepal_climate_data/nepal_admin_regions.gpkg')

# Display first few rows
print(admin_regions.head())

# Rename provinces for consistency
admin_regions['Province'] = admin_regions['name'].replace({
    'Bagmati': 'Bagmati Province',
    'Province 1': 'Koshi Province',
    'Province 2': 'Madhesh Province',
    'Gandaki': 'Gandaki Province',
    'Lumbini': 'Lumbini Province',
    'Karnali': 'Karnali Province',
    'Sudurpashchim': 'Sudurpashchim Province'
})

# Check available columns
print(admin_regions.columns)

# Verify coordinate reference system
print("admin_regions CRS:", admin_regions.crs)
print("nepal_map CRS:", nepal_map.crs)
```

# Merge GIS Data with Climate Vulnerability Index

```python
# Merge climate vulnerability data with Nepal map
temp = admin_regions.copy()
temp.drop(columns=['geometry'], inplace=True)

nepal_map = nepal_map.merge(temp, on='Province', how='left')

# Drop redundant column
nepal_map.drop(columns=['name'], inplace=True)

# Display merged data
print(nepal_map.head())
```

# Data Visualization (Climate Vulnerability)
### Climate Vulnerability Map (2020 vs 2050)


```python
fig, axes = plt.subplots(1, 2, figsize=(20, 10))

# 2020 Climate Vulnerability
nepal_map.plot(column='climate_vulnerability_2020', ax=axes[0], cmap='coolwarm', edgecolor='black', linewidth=0.2)
axes[0].set_title('Climate Vulnerability 2020')
axes[0].set_axis_off()

# 2050 Climate Vulnerability
nepal_map.plot(column='climate_vulnerability_2050', ax=axes[1], cmap='coolwarm', edgecolor='black', linewidth=0.2)
axes[1].set_title('Climate Vulnerability 2050')
axes[1].set_axis_off()

# Display plot
plt.tight_layout()
plt.show()
```

# Load and Analyze Glacier Data

### Glacier Data

```python
# Load Nepal glaciers data
nepal_glaciers = gpd.read_file('data/nepal_climate_data/nepal_glaciers.gpkg')

# Display first few rows
print(nepal_glaciers.head())
```

###  Visualize Glacier Retreat Trends

```python
# Histogram for glacier retreat in 2020
plt.figure(figsize=(10, 5))
sns.histplot(nepal_glaciers['retreat_2020'], bins=10, kde=True, color='blue')
plt.title('Distribution of Glacier Retreat in 2020')
plt.xlabel('Retreat (meters)')
plt.ylabel('Count')
plt.show()

# Scatter plot comparing glacier retreat (2020 vs 2050)
plt.figure(figsize=(8, 5))
sns.scatterplot(x=nepal_glaciers['retreat_2020'], y=nepal_glaciers['retreat_2050'])
plt.xlabel('Retreat in 2020')
plt.ylabel('Retreat in 2050')
plt.title('Glacier Retreat: 2020 vs 2050')
plt.show()
```

# Load and Analyze River Flow Data

```python
# Load Nepal rivers data
nepal_rivers = gpd.read_file('data/nepal_climate_data/nepal_rivers.gpkg')

# Display first few rows
print(nepal_rivers.head())


plt.figure(figsize=(10, 6))
sns.barplot(x='name', y='flow_2020', data=nepal_rivers, label='Flow 2020')
sns.barplot(x='name', y='flow_2050', data=nepal_rivers, label='Flow 2050', alpha=0.7)
plt.xlabel('River')
plt.ylabel('Flow')
plt.title('River Flow Comparison (2020 vs 2050)')
plt.legend()
plt.xticks(rotation=90)
plt.show()

```

# Load and Visualize Precipitation Data


```python
# Load precipitation data (GeoTIFF)
with rasterio.open("data/nepal_climate_data/nepal_precipitation_2020.tif") as src:
    precipitation_2020 = src.read(1)
    
with rasterio.open("data/nepal_climate_data/nepal_precipitation_2050.tif") as src:
    precipitation_2050 = src.read(1)

# Plot Precipitation Data
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
show(precipitation_2020, ax=ax[0], title="Precipitation 2020", cmap="inferno")
show(precipitation_2050, ax=ax[1], title="Precipitation 2050", cmap="inferno")
plt.show()
```

# Summary of Findings & Insights on Climate Change in Nepal (2020-2050)

## Key Findings:
Temperature Trends:

- Gradual increase in mean temperature (long-term warming trend).
Maximum temperature remains stable, indicating no extreme daytime heat rise.
Minimum temperature significantly increasing, leading to warmer nights.
Precipitation Patterns:

- Decline in mean precipitation in the eastern region.
Increase in extreme rainfall events in the Himalayas and Terai.
Rainfall redistribution, affecting water availability.
Glacier Retreat & River Flow:

- Accelerated glacier retreat affecting water sources.
Major rivers (Karnali, Mahakali, Koshi) showing reduced flow (12.5% drop in Karnali & Mahakali).
Climate Vulnerability:

- Province 2 is most vulnerable to climate change.
Sudurpashchim is least vulnerable but still at risk.

## Insights & Implications:
Ecosystem & Agriculture:

- Warmer nights impact biodiversity, crop cycles, and urban heat stress.
Farmers must adapt with drought-resistant crops & revised planting schedules.
Water & Disaster Management:

- Declining water availability due to glacier retreat & precipitation shifts.
Increased risk of floods & landslides, requiring stronger disaster preparedness.
Policy & Adaptation:

- Nepal needs sustainable water management & climate policies.
Region-specific strategies crucial to minimize uneven climate impacts.

# Conclusion:
Nepal is experiencing a long-term warming trend, glacier loss, and water resource stress. Urgent action in disaster preparedness, agriculture, and policy adaptation is needed to mitigate climate risks.
